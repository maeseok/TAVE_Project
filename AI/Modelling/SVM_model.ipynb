```python
# 상위 폴더 경로
parent_folder = '/content/남자배우'

# 상위 폴더 안의 모든 폴더를 순회
res = []
Act_List=os.listdir(parent_folder)
Act_List.sort()

for folder_name in Act_List:
    # 각각의 배우 폴더 경로 설정
    actor_folder = os.path.join(parent_folder, folder_name)

    # 배우 폴더 안의 사진 파일들의 경로 목록 가져오기
    photo_files = [os.path.join(actor_folder, file) for file in os.listdir(actor_folder)]
    photo_files.sort()
    folder_name=folder_name.split("_")
    # 사진 파일들을 로드하여 얼굴 및 이목구비 검출
    print(folder_name)
    face_landmarks_list = [folder_name[0]]
    print(face_landmarks_list)

    for photo_file in photo_files:
       try:
            # 이미지 로드
            image = face_recognition.load_image_file(photo_file)

            # 얼굴 검출
            face_locations = face_recognition.face_locations(image)

            if len(face_locations) > 0:  # 얼굴이 검출되었을 때만 처리
                # 얼굴 및 이목구비 검출
                face_landmarks = face_recognition.face_landmarks(image, face_locations=face_locations)

                if len(face_landmarks) > 0:  # 랜드마크가 검출되었을 때만 처리
                    # 31번 랜드마크 좌표값 가져오기
                    move_x, move_y = face_landmarks[0]['nose_bridge'][3]

                    # 랜드마크 조정
                    adjusted_landmarks = {}
                    for feature, landmarks in face_landmarks[0].items():
                        adjusted_feature = [(x - move_x, y - move_y) for x, y in landmarks]
                        adjusted_landmarks[feature] = adjusted_feature

            # 조정된 랜드마크를 리스트에 추가
            face_landmarks_list.append(adjusted_landmarks)
       except IsADirectoryError:
            pass

    res.append(face_landmarks_list)
```

```python
# 얼굴 랜드마크 정보를 .txt 파일에 저장 (로컬 경로 사용)
Data=[]
with open('output_file(m).txt', 'r') as file:
    Data = json.load(file)

print(Data)

X = []
y = []
for actor in Data : # 모든 배우 순회
    for i in range(1,len(actor)) : # 배우 내 이미지 수 만큼
        num = []
        for key, landmark in actor[i].items(): # 벡터값만 추출
            for tup in landmark:
                num.extend(tup)
        X.append(num)
        y.append(actor[0])
print(len(X))
print(y)
```

```python
fixed_X = []
for i in range(len(X)) :
    x_coords = np.array(X[i][::2])
    y_coords = np.array(X[i][1::2])
    '''center_x = np.mean(x_coords)
    center_y = np.mean(y_coords)'''
    fixed_x = x_coords
    fixed_y = y_coords

    fixed_X.append([coord for pair in zip(fixed_x, fixed_y) for coord in pair])
```

```python
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
#from sklearn.model_selection import GridSearchCV

# 학습 및 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(fixed_X, y, test_size=0.2, random_state=42)

# 특성 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

''' 하이퍼 파라미터 튜닝
param_grid = {'C': [0.1, 1, 10, 100],
              'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}

grid_search = GridSearchCV(svm.SVC(), param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)'''

# SVM 분류기 초기화
clf = svm.SVC(kernel='linear')
#clf = svm.SVC(**best_params)

# 모델 학습
#clf.fit(X_train_scaled, y_train)
clf.fit(X_train_scaled, y_train)

# 테스트 데이터로 예측
y_pred = clf.predict(X_test_scaled)

# 정확도 출력
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
